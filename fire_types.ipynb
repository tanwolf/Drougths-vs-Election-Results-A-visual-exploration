```python
# groupby auf state und year, Hektar werden pro Jahr gemittelt damit unterm Strich der selbe Wert erhalten bleibt; combined_BA wird über das ganze Jahr summiert
burn = df_burned_area_index.groupby(['State', 'Year']).agg({'Hectares': 'mean', 'combined_BA': 'sum', 'Cropland_BA_hectares':'sum', 'Forest_BA_hectares':'sum', 'Grass_and_Shrubland_BA_hectares':'sum', 'Wetlands_BA_hectares':'sum', 'Settlement_BA_hectares':'sum', 'Other_BA_hectares':'sum' }).reset_index()
```

```python
burn.head()
```

```python
# index wird errechnet als Prozent der gesamten Fläche, auf 2 Nachkommastellen gerundet, kann auch angepasst werden
burn['index'] = burn['combined_BA'] / burn['Hectares'] *100
```

```python
burn['Cropland_BA_hectares_index'] = burn['Cropland_BA_hectares'] / burn['Hectares'] *100
```

```python
burn['Forest_BA_hectares_index'] = burn['Forest_BA_hectares'] / burn['Hectares'] *100
```

```python
burn['Grass_and_Shrubland_BA_hectares_index'] = burn['Grass_and_Shrubland_BA_hectares'] / burn['Hectares'] *100
```

```python
burn['Wetlands_BA_hectares_index'] = burn['Wetlands_BA_hectares'] / burn['Hectares'] *100
```

```python
burn['Settlement_BA_hectares_index'] = burn['Settlement_BA_hectares'] / burn['Hectares'] *100
```

```python
burn['Other_BA_hectares_index'] = burn['Other_BA_hectares'] / burn['Hectares'] *100
```

```python
burn.sort_values(by='index')
```
