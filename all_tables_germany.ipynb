# join all tables

```python
import pandas as pd
```

```python
bundestag_election = pd.read_csv('Data/bundestag_election_2.csv')

try:
   droughtmagnitude = pd.read_csv('Data/droughtmagnitude_2000-2020.csv', encoding = "utf-8", sep=';')
except:
   droughtmagnitude = pd.read_csv('Data/droughtmagnitude_2000-2020.csv', encoding= "ISO-8859-1", sep=';')

burned_area = pd.read_csv('Data/burned_area_germany.csv', sep=';')
```

```python
all_tables = bundestag_election.merge(droughtmagnitude, how='left', left_on=['district_number', 'year'], right_on=['electoral_district_nr', 'Year'])
```

```python
all_tables.head(3)
```

```python
all_tables = all_tables.merge(burned_area, how='left', left_on=['federal_state_name','year'], right_on=['State','Year'])
```

```python
all_tables.info()
```

```python
all_tables.to_csv('Data/all_tables.csv', sep = ',', decimal=',', index =False)
```
