```python
df_burned_area= pd.read_csv('Data/df_burned_area.csv', sep=';')
```

```python
df_burned_area.head(3)
```

```python
df_burned_area.rename(columns={"Region": "State"}, inplace=True)
```

```python
df_burned_area.info()
```

```python
try:
    states_total_area_hectares = pd.read_csv('Data/us_states_total_area_hectares_corrected.csv', encoding = "utf-8", sep=';')
except:
    states_total_area_hectares = pd.read_csv('Data/us_states_total_area_hectares_corrected.csv', encoding= "ISO-8859-1", sep=';')
```

```python
states_total_area_hectares.head(3)
```

```python
states_total_area_hectares.info()
```

```python
new_df=states_total_area_hectares.merge(df_burned_area, on='State', how='left')
```

```python
new_df.head(3)
```

```python
new_df.info()
```

```python
new_df['burned_area_state_index']=new_df['combined_BA'] / new_df['Hectares'] * 100
```

```python
new_df.sort_values('burned_area_state_index', ascending=True).head(3)
```

```python
new_df.burned_area_state_index = new_df.burned_area_state_index.round()
```

```python
new_df.sort_values('burned_area_state_index', ascending=False).head(3)
```

```python
new_df.sort_values('burned_area_state_index', ascending=False).head(3)
```

```python
new_df['burned_area_state_index'].plot()
```

```python
new_df['Burned_Cropland_index'] = new_df['Cropland_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df['Burned_Forest_index'] = new_df['Forest_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df['Burned_Grass_and_Shrubland_index'] = new_df['Grass_and_Shrubland_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df['Burned_Wetlands_index'] = new_df['Wetlands_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df['Burned_Settlement_index'] = new_df['Settlement_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df['Burned_Other_index'] = new_df['Other_BA_hectares'] / new_df['Hectares'] * 100
```

```python
new_df.head(3)
```

```python
new_df.Burned_Cropland_index =new_df.Burned_Cropland_index.round()
```

```python
new_df.Burned_Forest_index =new_df.Burned_Forest_index.round()
```

```python
new_df.Burned_Grass_and_Shrubland_index =new_df.Burned_Grass_and_Shrubland_index.round()
```

```python
new_df.Burned_Wetlands_index =new_df.Burned_Wetlands_index.round()
```

```python
new_df.Burned_Settlement_index =new_df.Burned_Settlement_index.round()
```

```python
new_df.Burned_Other_index =new_df.Burned_Other_index.round()
```
