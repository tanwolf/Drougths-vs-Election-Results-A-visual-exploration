# Presidential Election Data Fine Tuning

```python
import pandas as pd
import numpy as np
```

```python
df_presedential_fine_tuning = pd.read_csv('Data/df_county_presidential.csv')
```

```python
df_presedential_fine_tuning.head(3)
```

```python
df_presedential_fine_tuning['year_short'] = df_presedential_fine_tuning['year'].str[-4:]
```

```python
df_presedential_fine_tuning.head()
```

```python
df_presedential_fine_tuning.info()
```

```python
df_presedential_fine_tuning['year_short'] = df_presedential_fine_tuning['year_short'].str.replace('0002','2000')
```

```python
df_presedential_fine_tuning.head(3)
```

```python
df_presedential_fine_tuning.info()
```

```python
for i in range(len(df_presedential_fine_tuning)):
  if df_presedential_fine_tuning.loc[i,'year'].endswith('.000002'):
    df_presedential_fine_tuning.loc[i,'year'] = df_presedential_fine_tuning.loc[i,'year'].replace('.000002', '.000002000')
```

```python
df_presedential_fine_tuning.head(3)
```

```python
df_presedential_fine_tuning.groupby(['county', 'year', 'party'])['candidate_votes'].max()
```

```python
df_presedential_fine_tuning.head(3)
```

```python
group = df_presedential_fine_tuning.sort_values(by='candidate_votes', ascending=False).groupby(['state','county', 'year_short']).agg({'candidate_votes':'max', 'party':'first', 'county_fips':'first', 'year' : 'first'}).reset_index()
```

```python
group.head(3)
```

```python
group[group['party']=='OTHER']
```

```python
group.county_fips = group.county_fips = group.county_fips.astype(str).str.rjust(5, '0')
```

```python
group.info()
```

```python
df_presedential_fine_tuning.info()
```

```python
group.to_csv('Data/df_county_presidential_agg.csv', sep = ',', index =False)
```
