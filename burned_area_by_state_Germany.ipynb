# Burned Area by State

```python
import pandas as pd
```

```python
df_burned_area = pd.read_csv('Data/MCD64A1_burned_area_full_dataset_2002-2019.csv')
```

```python
df_burned_area.head()
```

```python
df_burned_area.info()
```

```python
# Add new column (int) that sums all burned acres columns
df_burned_area.insert(12, 'combined_BA', int )
```

```python
df_burned_area.head()
```

```python
for i in range(len(df_burned_area)):
    df_burned_area.loc[i, 'combined_BA'] = df_burned_area.loc[i, 'Cropland_BA_hectares'] + df_burned_area.loc[i, 'Forest_BA_hectares'] + df_burned_area.loc[i, 'Grass_and_Shrubland_BA_hectares'] + df_burned_area.loc[i, 'Wetlands_BA_hectares'] + df_burned_area.loc[i,'Settlement_BA_hectares'] + df_burned_area.loc[i, 'Other_BA_hectares']
```

```python
df_burned_area.head()
```

```python
df_burned_area = df_burned_area[df_burned_area.GID_0 == 'DEU']
```

```python
df_burned_area.info()
```

```python
df_burned_area.head()
```

```python
df_burned_area= pd.read_csv('Data/df_burned_area.csv', sep=';')
```

```python
df_burned_area.head(3)
```

```python
df_burned_area.rename(columns={"Region": "State"}, inplace=True)
```

```python
df_burned_area.info()
```

```python
df_burned_area = df_burned_area.groupby(['State', 'Year']).agg({'combined_BA': 'sum', 'Cropland_BA_hectares':'sum', 'Forest_BA_hectares':'sum', 'Grass_and_Shrubland_BA_hectares':'sum', 'Wetlands_BA_hectares':'sum', 'Settlement_BA_hectares':'sum', 'Other_BA_hectares':'sum'}).reset_index()
```
